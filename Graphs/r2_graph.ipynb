```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
categories = ['GTPase KRas', 'G12C', 'G12D']
values1 = [0.663, 0.934, 0.676]  # Values for first set of bars
values2 = [0.663, 0.939, 0.642]  # Values for second set of bars

# Set the width of each bar
bar_width = 0.35

# Set the positions of the bars on the x-axis
x = np.arange(len(categories))

# Create the figure and axis
fig, ax = plt.subplots(figsize=(8, 6))

# Plot the bars
bar1 = ax.bar(x - bar_width/2, values1, bar_width, label='Neural '
                                                         'Network', 
              color='steelblue')
bar2 = ax.bar(x + bar_width/2, values2, bar_width, label='Random '
                                                         'Forest', 
              color='tomato')

# Add labels, title, and legend
ax.set_xlabel('Categories')
ax.set_ylabel('R2 score')
ax.set_title('R2 score for NN and RF')
ax.set_xticks(x)
ax.set_xticklabels(categories)
ax.legend()
ax.set_ylim(0, 1.0)
# Add value labels on top of each bar
ax.bar_label(bar1, padding=3)
ax.bar_label(bar2, padding=3)

# Adjust layout to prevent clipping of labels
plt.tight_layout()

# Show the plot
plt.show()
```
